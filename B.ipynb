```python
import sqlite3
import sys

import pandas as pd


def load_data():
    conn = sqlite3.connect(":memory:")
    cur = conn.cursor()

    with open("create_tables.sql", encoding="utf-8") as file:
        cur.executescript(file.read())

    return conn, cur
```

### Pregunta #1

```python
def test_01():
    conn, _ = load_data()
    query = """
    SELECT SUM(c12) FROM tbl1
    """
    return pd.read_sql_query(query, conn)

print(test_01())
```

### Pregunta #14

```python
--  Escriba una consulta que retorne por cada letra de la
--  columna K0 de la tabla tbl1, el valor promedio de la 
--  columna c21 en la tabla tbl2, para aquellos registros
--  con c13 mayor a 400.
```

```python
conn, _ = load_data()
query = """
SELECT * FROM tbl1
LIMIT 3
"""
pd.read_sql_query(query, conn)
```

```python
conn, _ = load_data()
query = """
SELECT *  FROM tbl2
LIMIT 3
"""
pd.read_sql_query(query, conn)
```

```python
conn, _ = load_data()
query = """
SELECT tbl1.k0, avg(c21)
FROM tbl1 INNER JOIN tbl2 ON tbl1.k1 = tbl2.k1
WHERE tbl1.c13 > 400 
GROUP BY tbl1.k0
"""
pd.read_sql_query(query, conn)
```
